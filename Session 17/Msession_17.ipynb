```python
df
```

```python
# how many super over finishes have occured
s_o = ipl[ipl['SuperOver'] == 'Y']['ID']
s_o.shape[0]
```

```python
ipl.head()
```

```python
# how many matches has csk won in kolkata

wins = ipl[(ipl['City']== 'Kolkata') & (ipl['WinningTeam'] ==  'Chennai Super Kings')]

wins.shape[0]
```

```python
# toss winner is match winner in percentage

(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0] * 100)/ipl.shape[0]
```
